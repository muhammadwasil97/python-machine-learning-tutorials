```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 10, 50)
```

```python
y = np.random.randint(1,10, 50)
y
```

```python
plt.plot(x)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(121)
ax.margins(10,10)
ax.plot(x,y,linewidth=2)
ax.set(xlim=(0, 10), xticks=np.arange(1, 11),
       ylim=(0, 10), yticks=np.arange(1, 11))
ax.set_title('Random XY Plot')

ax2 = plt.subplot(122)
ax2.margins(0.5,0.5)
y2 = 2 + np.sin(x)
ax2.plot(x, y2)

plt.show()
```
