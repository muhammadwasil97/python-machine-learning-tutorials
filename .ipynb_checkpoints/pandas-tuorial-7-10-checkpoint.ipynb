```python
# typecasting string values to numeric values
odi_records[['Inns','NO','Runs','HS','Ave','BF','SR','Century','Half Century','Ducks']] = odi_records[['Inns','NO','Runs','HS','Ave','BF','SR','100','50','0']].apply(lambda x : x.replace('-', '0'))
odi_records[['Inns','NO','Runs','HS','Ave','BF','SR','Century','Half Century','Ducks']] = odi_records[['Inns','NO','Runs','HS','Ave','BF','SR','Century','Half Century','Ducks']].apply(lambda x : x.replace('*', '')).apply(pd.to_numeric)
filtered_records = odi_records
filtered_records.dtypes
```

```python
filtered_records.describe()
```

```python
#tutorial 9
filtered_records.drop(['100', '50', '0'], axis=1, inplace=True)
filtered_records
```

```python
#tutorial 9
filtered_records.drop(['Unnamed: 0'], axis=1, inplace=True)
filtered_records
```

```python
debut_year = filtered_records['Span'].str.split('-').str.get(0)
retirement_year = filtered_records['Span'].str.split('-').str.get(1)
retirement_year
```

```python
filtered_records.insert(1, 'Debut Year', value=debut_year)
filtered_records.insert(2, 'Retirement Year', value=retirement_year)
```

```python
filtered_records
```

```python
# tutorial 10
filtered_records[['Debut Year','Retirement Year']] = filtered_records[['Debut Year','Retirement Year']].apply(pd.to_numeric)
```
