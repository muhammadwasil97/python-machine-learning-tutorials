```python
def sinplot(n=10, flip=1):
    x = np.linspace(0,14,100)
    for i in range(1, n+1):
        plt.plot(x, np.sin(x+i*0.5)*(n+2-i)*flip)
sinplot()
```

```python
sns.set_theme()
sns.set_style("whitegrid", {"axes.facecolor":".9"})
sns.set_context("notebook",font_scale=1.5, rc={"lines.linewidth":3})
sinplot(14)
sns.despine(offset=8, trim=True, left=True)
```

```python
sns.set_style("darkgrid", {"axes.facecolor":"0.2"})
sns.set_context("notebook",font_scale=1.5, rc={"lines.linewidth":3})
sinplot(n=20, flip=-1)
sns.despine(offset=8, trim=True, left=True)
```

```python
import math
def binaryplot():
    x = np.linspace(0,10, 1000)
    y = np.array([1 if math.floor(2*t)%2==0 else 0 for t in x])
    plt.plot(x,y)
binaryplot()
```

```python
f = plt.figure(figsize=(6,6))
gs = f.add_gridspec(2,2)

with sns.axes_style(style='whitegrid', rc={"axes.facecolor":".9"}):
    sns.set_context("notebook",font_scale=1.5, rc={"lines.linewidth":3})
    ax = f.add_subplot(gs[0,0])
    sinplot(6)
    
with sns.axes_style(style='darkgrid', rc={"axes.facecolor":".3"}):
    sns.set_context("notebook",font_scale=1.5, rc={"lines.linewidth":2})
    ax = f.add_subplot(gs[0,1])
    sinplot(14)

with sns.axes_style(style='whitegrid', rc={"axes.facecolor":".9"}):
    sns.set_context("notebook",font_scale=1.5, rc={"lines.linewidth":2})
    ax = f.add_subplot(gs[1,0])
    binaryplot()
    
f.tight_layout()
```
