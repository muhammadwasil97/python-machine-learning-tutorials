```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 10, 50)
```

```python
y = np.random.randint(1,10, 50)
y
```

```python
plt.plot(x)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
# fig = plt.figure()
ax = plt.subplot(221)
ax.margins(10,10)
ax.plot(x,y,linewidth=2)
ax.set(xlim=(0, 10), xticks=np.arange(1, 11),
       ylim=(0, 10), yticks=np.arange(1, 11))
ax.set_title('Random XY Plot')

ax2 = plt.subplot(222)
ax2.margins(0.5,0.5)
y2 = 2 + np.sin(x)
ax2.plot(x, y2)

ax3 = plt.subplot(223)
# ax3.margins(1.5,1.5)
x3 = np.linspace(1,10, 100)
y3 = 4 + np.sin(2*x3)
ax3.plot(x3, y3)

plt.show()
```
